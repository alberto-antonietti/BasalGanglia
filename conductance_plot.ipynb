```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.stats as stats
import math
import seaborn as sns
from scipy.stats import truncnorm
```

```python
# user input
myclip_a = 2.4e-2
myclip_b = 6.4e-1
my_mean = 2.4e-1
my_std = 1e-1

a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std
x_range = np.linspace(-1e-2,1,1000)

fs_clip_a = 1.1e-1
fs_clip_b = 6.1
fs_mean = 1.1
fs_std = 1.5

fs_a, fs_b = (fs_clip_a - fs_mean) / fs_std, (fs_clip_b - fs_mean) / fs_std
fs_x_range = np.linspace(-1e-2,6,1000)

y_range = np.linspace(0,4,1000)

fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.axes.yaxis.set_visible(False)

plt.plot(x_range, truncnorm.pdf(x_range, a, b, loc = my_mean, scale = my_std), color = 'g', label = 'dSPN or iSPN')
plt.plot(fs_x_range, truncnorm.pdf(fs_x_range, fs_a, fs_b, loc = fs_mean, scale = fs_std), color = 'orange', label = 'FS')
plt.vlines(x=3, ymin=0, ymax=4, color = 'red', label = 'LTS')
plt.legend(loc="upper right", title = 'Pre-Synaptic Cell Type')

ax2.set_xlabel('Conductance [nS]')
ax2.set_title("Synapse Weight Distribution")
plt.show()
```

```python
fs_clip_a = 1.1e-10
fs_clip_b = 6.1e-09

fs_mean = 1.1e-09
fs_std = 1.5e-09

fs_a, fs_b = (fs_clip_a - fs_mean) / fs_std, (fs_clip_b - fs_mean) / fs_std
fs_x_range = np.linspace(-1e-11,6e-9,1000)

fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.axes.yaxis.set_visible(False)
plt.plot(fs_x_range, truncnorm.pdf(fs_x_range, fs_a, fs_b, loc = fs_mean, scale = fs_std), color = 'orange')
ax2.set_xlabel('Conductance [S]')
ax2.set_title("Synapse Weight Distribution - Fast Spiking Neurons (FS)")
plt.show()
```
