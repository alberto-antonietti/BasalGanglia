```python
def plotSpikes(events, id_array, color, label, ax):
    
    label_done = False
    
    for i in id_array:
        
        spikes = events[i] #array of the spike times of neuron i
        if not label_done:
            ax.plot(spikes, np.full_like(spikes, i), marker=".", label=label, color=color, linestyle="None")
            #this full_like function generates an array that has the same
            #size as events[i], with the value i on every position (so we
            #have the same number of x and ys for plotting)
            label_done = True
        else:
            ax.plot(spikes, np.full_like(spikes, i), marker=".", color=color, linestyle="None")
    return();

def plotRaster(folder_path):
    ''' Given a path to a folder inside NEURON-data, generates the raster plot for the corresponding simulation
        (folder_path is "NEURON-data/net_X" )
    '''    
    spikes_dict = getSpikes(folder_path,100)
    
    n_ids = np.array(list(spikes_dict.keys()), dtype=int)
    n_ids.sort() #organize the neurons by id

    neuron_matrix = getNeuronMatrix(folder_path)
    dspn_ID, ispn_ID, lts_ID, fs_ID, chin_ID = getNeuronSubMatrixes(neuron_matrix)

    pop_range1 =  range(0, 50)
    pop_range2 =  range(51, 100)

    fig_handle = plt.figure()
    ax = fig_handle.add_subplot(111)
    ax.set_xlabel('$t$ (ms)')

    plotSpikes(spikes_dict, fs_ID, 'b', 'fs', ax)
    plotSpikes(spikes_dict, dspn_ID, 'c', 'dspn', ax)
    plotSpikes(spikes_dict, lts_ID, 'black', 'lts', ax)
    plotSpikes(spikes_dict, ispn_ID, 'm', 'ispn', ax)
    plotSpikes(spikes_dict, chin_ID, 'orange', 'chin', ax)


    plt.legend()
    plt.show()
```

```python
plotRaster('NEURON-data/net_100')
```
